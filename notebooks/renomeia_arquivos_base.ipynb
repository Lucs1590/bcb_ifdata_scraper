```python
import os
import shutil
```

```python
download_folder = os.path.join('downloads')
folders = [
    'conglomerados_prudenciais',
    'conglomerados_financeiros',
    'instituicoes_individuais',
    ''
]
```

```python
for folder in folders:
    path_folder = os.path.join(download_folder, folder)
    print(path_folder)
    
    for file_name in os.listdir(path_folder):
        file_path = os.path.join(path_folder, file_name)
        
        if not file_path.endswith('.csv'):
            continue
        
        data_base = file_name[:6]
        trimestre = data_base[0:2]
        ano = data_base[2:]
        restante_arquivo = file_name[6:]

        file_name_new = '{}_{}{}'.format(ano, trimestre, restante_arquivo)
        file_path_new = os.path.join(path_folder, file_name_new)

        print('Renomeando {}'format(file_path))
        shutil.move(file_path, file_path_new)
```
