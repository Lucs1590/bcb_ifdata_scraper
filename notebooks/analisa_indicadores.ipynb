```python
import pandas as pd
import os
```

```python
def merge_relatorios_df(tipo_if):
    # lê os arquivos csv
    file_name = '{}_ativo_import.csv'.format(tipo_if)
    file_path = os.path.join('bases_import', file_name)
    df1 = pd.read_csv(file_path, sep=";", low_memory=False)
    
    file_name = '{}_demonstracao_resultado_import.csv'.format(tipo_if)
    file_path = os.path.join('bases_import', file_name)
    df2 = pd.read_csv(file_path, sep=";", low_memory=False)

    file_name = '{}_passivo_import.csv'.format(tipo_if)
    file_path = os.path.join('bases_import', file_name)
    df3 = pd.read_csv(file_path, sep=";", low_memory=False)

    file_name = '{}_resumo_import.csv'.format(tipo_if)
    file_path = os.path.join('bases_import', file_name)
    df4 = pd.read_csv(file_path, sep=";", low_memory=False)

    # Merging the dataframes                       
    cols_to_use = df2.columns.difference(df1.columns)
    cols_to_use = cols_to_use.to_list() + ['co_if', 'dt_base']
    df5 = pd.merge(df1, df2[cols_to_use], how='inner', on=['co_if', 'dt_base'])

    # Merging the dataframes
    cols_to_use = df4.columns.difference(df3.columns)
    cols_to_use = cols_to_use.to_list() + ['co_if', 'dt_base']
    df6 = pd.merge(df3, df4[cols_to_use], how='inner', on=['co_if', 'dt_base'])

    # Mergind the final dataframe
    cols_to_use = df6.columns.difference(df5.columns)
    cols_to_use = cols_to_use.to_list() + ['co_if', 'dt_base']
    df = pd.merge(df5, df6[cols_to_use], how='inner', on=['co_if', 'dt_base'])

    return df
```
